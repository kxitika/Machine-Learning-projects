```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
from sklearn.datasets import load_diabetes
```

```python
dataset = load_diabetes()
```

```python
print(dataset['DESCR'])
```

```python
#dataset.data
```

```python
#independent features
df_diabetes = pd.DataFrame(data=dataset.data, columns = ['age','sex','bmi','bp','s1','s2','s3','s4','s5','s6'])
```

```python
df_diabetes
```

```python
y = dataset['target']
```

```python
y
```

```python
X = df_diabetes
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state = 10)
```

```python
X_train.corr() #gives correlation b/w different independent features
```

```python
plt.figure(figsize = (15,10))
sns.heatmap(X_train.corr(), annot = True)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(X_train,y_train)
```

## Hyperparameter tuning

```python

param = {
    'criterion' : ['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
    'splitter' : ['best', 'random'],
    'max_depth' : [1,2,3,4,5,10,15,20,25],
    'max_features' : ['sqrt', 'log2']

}
          
```

```python
from sklearn.model_selection import GridSearchCV
regressor = DecisionTreeRegressor()
grid = GridSearchCV(estimator=regressor, param_grid = param, scoring = 'neg_mean_squared_error', cv = 5)
```

```python
# to remove the warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

```python
model = DecisionTreeRegressor(criterion='friedman_mse', max_depth=3, max_features='sqrt',splitter='best')
model.fit(X_train,y_train)
```

```python
#visualize the decision tree
from sklearn import tree
plt.figure(figsize = (15,10))
tree.plot_tree(model, filled = True)
```
